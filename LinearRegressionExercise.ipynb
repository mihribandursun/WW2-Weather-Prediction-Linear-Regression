```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
X_train.shape
```

```python
def correlation_for_dropping(df, threshold):
    columns_to_drop = set() #Silinecek sütun isimlerini saklamak için bir küme (set) oluşturulur.
    corr = df.corr() #korelasyonu hesaplar
    for i in range(len(corr.columns)):
        for j in range(i): # değerlerin kendileriyle ve aynı değerlerle karşılaştırılmasını engelliyoruz
            if abs(corr.iloc[i,j]) > threshold: #İki sütun arasındaki ilişkinin mutlak değerine bakar.
                columns_to_drop.add(corr.columns[i]) #yüksek korelasyonlu olan çiftlerden birini "silinecekler" listesine ekler
    return columns_to_drop
```

```python
columns_dropping = correlation_for_dropping(X_train, 0.85)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X_train")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("X_train_scaled")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
linear = LinearRegression()
linear.fit(X_train_scaled, y_train)
y_pred = linear.predict(X_test_scaled)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean Absolute Error: ",mae)
print("Mean Squared Error: ",mse)
print("R2 Score: ",score)
```
